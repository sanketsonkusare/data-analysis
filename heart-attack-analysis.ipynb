**Installing packages**

```r
install.packages("tidyverse")
install.packages("ggplot2")
library ("tidyverse")
library ("ggplot2")
```

**Importing data**

```r
heart <- read.csv("../input/heart-dataset/heart.csv")
```

**Viewing the data**

```r
View (heart)
```

**Creating scattered plot**

```r
ggplot(data = heart) + geom_point(mapping = aes(x=chol, y=age,color=sex))+
  facet_wrap(~sex) + labs(title = "Heart details of male/female and Cholesterol")
```

data = heart selects the data to make plot of
'+' is added to add a layer in the plot 
geom_point is used to make point in the plot here geom is geometric object and point is used to make point in the plot
color is used to differentiate between the two different categories falls in the same plot
here facet_wrap is used to make different plots according to selected category

**creating bar graph**

```r
ggplot(data = heart) + geom_bar(mapping = aes(x=cp))
```

**Creating pie chart**

```r
x <- heart[1:12,"oldpeak"]
labels <- heart[1:12,"oldpeak"]
pie(x,labels,main="old peak heart disease",col=rainbow(length(x)))
```

**Creating box plot**

```r
ggplot(data = heart) + geom_boxplot (mapping = aes(x=age,y=cp))
```

**Creating line graph**

```r
ggplot(data = heart) + geom_smooth(mapping = aes(x=age,y=chol))
```
